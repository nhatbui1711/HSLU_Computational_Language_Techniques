```python
import pandas as pd
import ast
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from hdbscan import HDBSCAN
from umap import UMAP
```

```python
df = pd.read_csv('/content/drive/MyDrive/CTL_Project/preprocessed_data.csv')
```

```python
df.head()
```

```python
df['content_tokens_lemmatized'] = df['content_tokens_lemmatized'].astype(str)
documents = df['content_tokens_lemmatized'].apply(lambda x: ' '.join(ast.literal_eval(x)))
docs = documents.to_list()
print(docs[0])
```

```python
hdbscan_model = HDBSCAN(min_cluster_size=10, metric='euclidean', cluster_selection_method='eom', prediction_data=True)
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
# Create a UMAP model with a higher n_neighbors value
# umap_model = UMAP(n_neighbors=50, n_components=5, min_dist=0.0, metric='cosine')


topic_model = BERTopic(
    embedding_model=embedding_model,
    hdbscan_model=hdbscan_model,
    # umap_model=umap_model,
    verbose=True)
```

```python
# Train the model and identify topics
topics, probabilities = topic_model.fit_transform(docs)
```

```python
# Get a DataFrame with information about each topic
topic_info = topic_model.get_topic_info()
print(topic_info)
```

```python
# Visualize the topics
# This creates an interactive plot where you can see the topics and their relationships.
topic_model.visualize_topics()
```

```python
# Visualize the topic hierarchy
# This is useful for understanding how topics might be related to each other.
topic_model.visualize_hierarchy()
```

```python
# Create a bar chart of the most frequent words for a few topics
topic_model.visualize_barchart(top_n_topics=10)
```
